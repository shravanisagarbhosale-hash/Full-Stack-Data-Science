```python
# 1. Deep Indexing Practice Given the list: data = ["a", "b", ["c", ["d", "e", "target"], "f"], "g"]
# •	Task: Write the code to extract the word "target".
data = ["a", "b", ["c", ["d", "e", "target"], "f"], "g"]
data[2][1][2]

#
#data[2] extracts the inner list at index 2:["c", ["d", "e", "target"], "f"]
#data[2][1] extracts the further nested list at index 1 of that sublist:["d", "e", "target"]
#data[2][1][2] extracts the element at index 2 of the final sublist:"target"
```

```python
# 2. Append vs. Extend Given: colors = ["red", "blue"]
#•	Task A: Add the list ["green", "yellow"] so the result is ['red', 'blue', ['green', 'yellow']].
#•	Task B: Add the same list so the result is ['red', 'blue', 'green', 'yellow'].
colors = ["red", "blue"]
print("Append ")
colors.append(["green", "yellow"])
colors
```

```python
colors_extend = ["red", "blue"]
print("Extend ")
colors_extend.append(["green", "yellow"])
colors_extend
```

```python
# 3. Matrix Access Matrices are often represented as nested lists: matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
# •	Task: Access the number 6.
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix[1][2]
```

```python
# 4. Using .insert() Given: nums = [10, 20, 40]
# •	Task: Insert the number 30 at the correct spot to keep the list numerical.
nums = [10, 20, 40]
nums.insert(2, 30)
nums
```

```python
# 5. Mixed Data Types Given: mixed = [1, "Python", [True, False]]
# •	Task: Change False to None.
mixed = [1, "Python", [True, False]]
mixed[2][1] = "None"
mixed
```

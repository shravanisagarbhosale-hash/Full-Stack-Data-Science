```python
# Area Calculator: Write a function area_rect(length, width) that calculates and prints the area.
def area_rect(length, width):
    area = 0
    area = length * width
    print(f"Area of rectangle having sides {length} and {width} is {area}")

area_rect(5, 9)
```

```python
# Voting Eligibility: Write a function can_vote(age) that prints "Eligible" if age is 18 or older, and "Not Eligible" otherwise.
def can_vote(age):
    if age >= 18:
        print("Eligible")
    else:
        print("Not Eligible")

can_vote(20)
can_vote(15)
```

```python
# Max Number: Write a function find_max(a, b) that prints the larger of the two numbers.
def find_max(a, b):
    if a > b:
        print(f"{a} is greater than {b}")
    else:
        print(f"{b} is greater than {a}")

find_max(100, 500)
```

```python
# String Length: Create a function check_len(s) that prints whether a string is "Long" (more than 10 chars) or "Short".
def check_len(s):
    length = len(s)
    if length > 10:
        print("Entered string is long")
    else:
        print("Entered string is short")

check_len("Hello")
check_len("Hello, This is an example of function")
```

```python
# Temperature Converter: Write a function that takes Celsius and prints the Fahrenheit equivalent.
def cel_fah(celsius):
    fahrenheit = (celsius * 9/5) + 32
    print(f"Celsius = {celsius} -> Fahrenheit = {fahrenheit}")

cel_fah(45)
```

### Part C: Loops & Functions (Level 3)

```python
# Table Generator: Write a function make_table(n) that prints the multiplication table of n from 1 to 10 using a for loop.
def make_table(n):
    print("Table of ", n, " is : ")
    for i in range(1, 11):
        multiply = i * n
        print(f"{n} x {i} = {multiply}")

make_table(18)
```

```python
# List Sum: Write a function that takes a list of numbers and prints their total sum.
def num(numbers):
    total = sum(numbers)
    print("Total addition is ", total)

mylist = [1, 2, 3, 4, 65]
num(mylist)
```

```python
# Vowel Counter: Create a function count_vowels(word) that loops through a string and prints the number of vowels found.
def count_vowels(word):
    vowels = 'aeiou'
    count = 0
    for char in word.lower():
        if char in vowels:
            count += 1
    print(f"The number of vowels in '{word}' is: {count}")

count_vowels("Python Programming")
```

```python
# Countdown: Write a function that takes a number x and uses a while loop to count down to zero.
def number(x):
    while x >= 0:
        print(x)
        x -= 1

number(10)
```

```python
# Filtering: Write a function that takes a list of numbers and prints only the ones that are divisible by 3.
def divisible_by_3(number_list):
    for i in number_list:
        if i % 3 == 0:
            print(f"{i} is divisible by 3")

user_input = input("Enter numbers seperated by space in between : ")
new_list = [int(x) for x in user_input.split()]

divisible_by_3(new_list)
```

### Part D: Expert "Top 1%" Practices (Level 4)

```python
# Default Arguments: Create a function power(base, exp=2) so that if no exponent is provided, it automatically squares the number.
def power(base, exp=2):
    print(f"{base}^{exp} = ", base ** exp)

power(15, 3)
power(16)
```

```python
# Simple Calculator: Create a function calc(a, b, operation) where operation can be "add", "sub", "mul", or "div".
def calc(a, b, operation):
    if operation.lower() == 'add' or operation == '+':
        add = a + b
        print(f"Addition of {a} and {b} is {add}")
    elif operation.lower() == 'sub' or operation == '-':
        sub = a - b
        print(f"Subtraction of {a} and {b} is {sub}")
    elif operation.lower() == 'mul' or operation == '*':
        mul = a * b
        print(f"Multiplication of {a} and {b} is {mul}")
    elif operation.lower() == 'div' or operation == '/':
        div = a / b
        print(f"Division of {a} and {b} is {div}")
    else:
        print("Please select a correct operation (+,-,*/). Try again!")

calc(56, 36, '+')
calc(56, 36, '-')
calc(6, 6, '*')
calc(36, 4, '/')
```

```python
# Password Validator: Write a function that checks if a password is longer than 8 characters AND contains the symbol "@".
def check_password(password):
    if len(password) >= 8 and '@' in password:
        print("Password is valid")
    else:
        print("Password is not valid")

check_password('abcs')
check_password('abcd12345')
check_password('abcd@12345')
```

```python
# Nested Logic: Write a function that takes a list of ages and uses the even_odd logic from your notebook to count how many even-numbered ages exist.
def even_odd(num):
    if num % 2 == 0:
        return True
    else:
        return False

def check_age(age_list):
    print("Even-numbered ages are : ")
    for i in age_list:
        if even_odd(i) == True:
            print(i)
    else:
        return

user_input = input("Enter numbers seperated by space in between : ")
new_age_list = [int(x) for x in user_input.split()]

check_age(new_age_list)
```

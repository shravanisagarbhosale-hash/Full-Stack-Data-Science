# 3. While Loop Practice Questions

### Basic Level

```python
# 1. Countdown Timer:
# Write a program to count down from a given number to 0 using a while loop.

num = int(input("Enter a number "))
while num >= 0:
    print(num)
    num -= 1
```

```python
# 2. Guess the Number:
# Write a guessing game where the user keeps guessing a predefined number until they get it
# right.

num = int(input("Enter any number "))
guessedno = 10

while num != guessedno:
    print("Your guess is wrong. Try again. ")
    num = int(input("Enter any number: "))
print("Hurray ! You have guessed the number right. ", )
```

```python
# 3. Simple Sum Calculator:
# Keep asking the user for numbers and calculate the sum until they enter 0.

num1 = int(input("Enter first number : "))
num2 = int(input("Enter second number : "))
sumNos = 0;

while (num1 != 0) and (num2 != 0):
    sumNos = num1 + num2
    print(f"The sum of {num1} and {num2} is {sumNos}.")
    num1 = int(input("Enter first number : "))
    num2 = int(input("Enter second number : "))

print("You have entered 0. Try again using another number.")
```

### Intermediate Level

```python
# 4. Reverse Digits:
# Input a number and reverse its digits using a while loop.
# Example: 12345 → 54321.

num = abs(int(input("Enter any number ")))
print("Original number is : ", num)
digit = 0

while num > 0:
    temp = num % 10
    digit = (digit * 10) + temp
    num = num // 10

print("Reversed number is ", digit)
```

```python
# 5. Find the Smallest Divisor:
# Write a program to find the smallest divisor of a given number other than 1.

num = int(input("Enter a positive number "))
divisor = 2

while num % divisor != 0:
    divisor += 1

print(f"The smallest divisor of {num} is : {divisor}")
```

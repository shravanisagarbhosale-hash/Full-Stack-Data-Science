```python
# 6. Word Counter: Read a text file and count how many total words are in it.
f = open('bio.txt', mode='r+')
count_words = f.read().split(" ")
word_count = len(count_words)
print("Total words = ", word_count)
```

```python
# 7. Line Finder: Read a file and print only the lines that contain the word "Python".
f = open('data.txt', mode='w')
f.write("""We are learning Python Programming. 
It is highy-interpreted, easy to learn and object oriented language. 
Python was invented by GUISSO VAN ROSSUM in 1990.""")
f.close()

f = open('data.txt', mode='r+')
for line in f:
    if "Python" in line:
        print(line.strip())
```

```python
# 8. Case Normalizer: Read a file, convert all text to lowercase, and save it into a new file called cleaned.txt.
f = open('data.txt', mode='r')
c = open('cleaned.txt', mode = 'w')
for line in f: 
    c.write(line.lower())
c.close()
f.close()
```

```python
c = open('cleaned.txt', mode = 'r')
c.seek(0)
c.read()
```

```python
# 9. Unique Words: Read a file and print a list of all unique words (hint: use a Set).
unique_words = set()
c = open('cleaned.txt', mode = 'r+')
for line in c:
    words = line.split()
    unique_words.update(words)

print(unique_words)
```

```python
# 10. Log Collector: Write a loop that takes 5 different inputs from a user and saves each one on a new line in a file.
new = open('logCollector.txt', 'w')
for i in range(1, 6):
    name = input("Enter any input : ")
    new.write(name + "\n")

new.close()
```

```python
# 11. File Merger: Read content from file1.txt and file2.txt, then combine them into merged.txt.
f1 = open('bio.txt', mode='r')
content1 = f1.read()

f2 = open('data.txt', mode='r')
content2 = f2.read()

f3 = open('merged.txt', mode='w')
f3.write(content1 + content2)

f3.close()
f2.close()
f1.close()
```

```python
# 12. Search and Replace: Write a script that finds every occurrence of the word "bad" in a file and replaces it with "good".
f = open("bad.txt", mode='w')
f.write("""It was a thoroughly bad Tuesday morning, starting when a bad alarm clock malfunction caused a bad delay for my morning commute. 

Traffic was in a bad state, creating a bad bottleneck that put me in a truly bad mood before the workday even began. 

My boss gave me some bad feedback on a project, which was a bad blow to my confidence. 

To top it off, I had a bad lunch that left a bad taste in my mouth for the rest of the afternoon
""")

f.close()
```

```python
f = open("bad.txt", mode='r')
f.read()
```

```python
good = open("good.txt", mode='w')
for line in f:
    line.replace('bad', 'good')

good.close()
```

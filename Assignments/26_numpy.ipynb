```python
# 1. The List Challenge: Recreate the time comparison from your notebook. Compare a Python list and a NumPy array of 1,000,000 elements for a 
# cubing operation ($x^3$). Record the time difference.

import time

elements = 1000000
python_list = list(range(elements))
np_array = np.arange(elements)

start_list = time.time()
list_cube = [x**3 for x in python_list]
end_list = time.time()
final_list_time = end_list - start_list 

start_array = time.time()
array_cube = np_array**3
end_array = time.time()
final_array_time = end_array - start_array

print(f"Total time taken for list is : {final_list_time} seconds")
print(f"Total time taken for array is : {final_array_time} seconds")
print(f"Difference is : {final_list_time - final_array_time} seconds")
print(f"NumPy is {final_list_time / final_array_time:.2f}x faster!")
```

```python
# 8. Normalization: Given a sales array, subtract the average from every element and divide by the standard deviation. This is a common "Scaling" 
# technique in Machine Learning.

products = ['Lapotp', 'Phones', 'SmartPhones', 'Computer']
sales_array = np.array([80, 50, 60, 56], dtype = float)
print(sales_array)

avg = np.mean(sales_array)
#avg = np.average(sales_array)

# standard deviation
sd = np.std(sales_array)

normalized = (sales_array - avg) / sd
print(f"Average : {avg:.2f}")
print(f"Standard Deviation : {sd:.2f}")
print(f"Final array is : {normalized}")
```

```python
# 9. Horizontal/Vertical Stacking: Research np.hstack() and np.vstack(). Combine two matrices together.
# In Python, np.hstack() and np.vstack() are NumPy library functions used to join multiple arrays together by stacking them along different axes.
# np.hstack() stands for horizontal stack; it joins arrays side-by-side (column-wise).
# np.vstack() stands for vertical stack; it joins arrays on top of each other (row-wise).

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[6, 7], [8, 9]])

horizontal = np.hstack((matrix1, matrix2))
vertical = np.vstack((matrix1, matrix2))

print("Horizontal Stack \n: ", horizontal)
print("\n\nVertical Stack \n: ", vertical)
```

```python
# 10. Finding Extremes: Find the index of the maximum value in an array using np.argmax().
# np.argmax() is a NumPy function that returns the index (position) of the largest value in an array.
matrix = np.array([[10, 20, 90], [95, 63, 78], [100, 56, 85]])
index = np.argmax(matrix)
print("Maximum value index is : ", index)
```

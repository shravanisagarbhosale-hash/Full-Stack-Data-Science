### Part C: Cleaning & Transformation (Level 3) 

```python
# 11. Smart Fill: Fill the missing "Age" values not with a single number, but with the median age of the entire dataset.
df['Age'].median()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
print(df['Age'].isna().sum())
```

```python
# 12. Categorical Swap: Change "male" to 1 and "female" to 0 in the "Sex" column using .map() or .replace().
def gender(x):
    if x == 'male':
        return 1
    else:
        return 0

df['New_Age'] = df['Sex'].apply(gender)
df.head()
```

```python
# 13. Null Drop: Drop all rows where "Cabin" is missing. How many rows are left?
new_Cabin = df.dropna(subset="Cabin")
new_Cabin
```

```python
print(len(new_Cabin))
```

```python
# 14. Sorting: Sort the DataFrame by "Fare" (highest first) and then by "Age" (youngest first).
multisort = df.sort_values(by = ["Fare", "Age"], ascending = [False, True])
print(multisort[["Name", "Age", "Fare"]].head())
```

```python
# 15. NLP Text Cleaning: In the "Name" column, remove all punctuation marks using .str.replace() and a regex.
df["Name"].str.replace(',', "")
```

```python
# 16. Group Analysis: Group the data by "Pclass" and calculate the mean fare for each class.
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv")
group = df.groupby('Pclass')['Fare'].mean()
group
```

```python
# 19. Merging: Create two DataFrames (one with names and IDs, another with IDs and Marks) and use pd.merge() to combine them.
data1 = {
    'Name' : ['Alice', 'George', 'Bob', 'John', 'Sia'],
    'ID' : [101, 102, 103, 104, 105]
}
df1 = pd.DataFrame(data1)

data2 = {
    'ID' : [101, 102, 103, 104, 105],
    'Marks' : [95, 96, 85, 45, 68]
}
df2 = pd.DataFrame(data2)

merged_dataset = pd.merge(df1, df2)
merged_dataset
```

### Part A: Single Inheritance (Level 1)

```python
# 1. Create a parent class Animal with a method eat(). Create a child class Dog that inherits Animal and adds a method bark().
class Animal:
    def eat(self):
        print("Animal eats food")

class Dog(Animal):
    def bark(self):
        print("Dog is barking.")

tommy = Dog()
tommy.eat()
tommy.bark()
```

```python
# 2. Create a class Vehicle with brand and year. Create a child class Car that adds fuel_type.
class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def info(self):
        print(f"Brand : {self.brand}\nYear : {self.year}")

class Car(Vehicle):
    def __init__(self, brand, year, fuel_type):
        super().__init__(brand, year)
        self.fuel_type = fuel_type

    def more_info(self):
        print(f"Fuel type : {self.fuel_type}")

punch = Car("Tata", 2023, "Petrol")
punch.info()
punch.more_info()
```

```python
# 3. Method Overriding: Create a parent class Shape with a method draw(). In the child class Square, rewrite the draw() method to print "Drawing a Square".
class Shape:
    def draw(self):
        print("Drawing a shape.......")

class Square(Shape):
    def draw(self):
        print("Drawing a Square......")

shape = Shape()
shape.draw()

sq = Square()
sq.draw()
```

```python
# 4. Write a class Person with name. Create a child Employee with employee_id.
class Person:
    def __init__(self, name):
        self.name = name

class Employee(Person):
    def __init__(self, name, employee_id):
        super().__init__(name)
        self.employee_id = employee_id

    def emp_info(self):
        return f"Name : {self.name}\nEmployee ID : {self.employee_id}"

emp = Employee("Shravani", 1405)
print(emp.emp_info())
```

```python
# 5. Check if an object is an instance of a specific class using the isinstance() function.
p = Person("Shreya")

print(f"Is emp an Employee :", isinstance(emp, Employee))
print(f"Is p an Employee :", isinstance(p, Person))
print(f"Is emp an Employee :", isinstance(emp, Person))
print(f"Is p an Employee :", isinstance(p, Employee))
```

### Part B: Multilevel Inheritance (Level 2)

```python
# 6. The Tech Chain: Create a class ElectronicDevice. Inherit it into Computer. Then inherit Computer into Laptop. Add one unique method to each.
class ElectronicDevice:
    def __init__(self, name):
        self.name = name

    def show_info(self):
        print(f"Hello, This is an Electronic Device class. I am a {self.name}")

class Computer(ElectronicDevice):
    def __init__(self, name, comp_type):
        super().__init__(name)
        self.comp_type = comp_type

    def show_computer_info(self):
        print(f"Hello, This is a Computer class. I am a '{self.comp_type}' type of computer.")

class Laptop(Computer):
    def __init__(self, name, comp_type, brand):
        super().__init__(name, comp_type)
        self.brand = brand

    def show_laptop_info(self):
        print(f"Hello, This is a Laptop class. I am from {self.brand} brand.")

mylaptop = Laptop("Office Device", "Personal Computer", "HP")
print(mylaptop.show_info())
print(mylaptop.show_computer_info())
print(mylaptop.show_laptop_info())
```

```python
# 7. Create a hierarchy: LivingBeing $\rightarrow$ Animal $\rightarrow$ Mammal. Ensure the Mammal object can access methods from LivingBeing.
class LivingBeing:
    def display(self):
        print("\nThe biological hierarchy goes from a broad living being (any organism), to an animal (a multicellular organism that moves and consumes organic matter), and finally to a mammal (a specific warm-blooded animal that has hair and nurses its young).")

class Animal(LivingBeing):
    def display_animal_info(self):
        print("\nAn animal is a living being made of multiple cells that breathes, eats organic matter to survive, and can move independently.")

class Mammal(Animal):
    def display_mammal_info(self):
        print("\nA mammal is a warm-blooded animal that has hair or fur and feeds its young with milk.")

cows = Mammal()
cows.display_mammal_info()
cows.display_animal_info()
cows.display()
```

```python
# 8. Initialization Chain: In a multilevel setup, use the constructor (__init__) in the top-level class and see if the bottom-level class can access 
# those attributes.
class TopLevelClass:
    def __init__(self, topAttribute):
        self.topAttribute = topAttribute

class MidLevelCLass(TopLevelClass):
    def __init__(self, topAttribute, midAttribute):
        super().__init__(topAttribute)
        self.midAttribute = midAttribute

class LowerLevelClass(MidLevelCLass):
    def __init__(self, topAttribute, midAttribute, lowAttribute):
        super().__init__(topAttribute, midAttribute)
        self.lowAttribute = lowAttribute

    def displayAllCLasses(self):
        print(f"Top Level Class : {self.topAttribute}")
        print(f"Mid Level Class : {self.midAttribute}")
        print(f"Low Level Class : {self.lowAttribute}")

attributes = LowerLevelClass("Top Attribute", "Mid Attribute", "Low Attribute")
attributes.displayAllCLasses()
```

```python
# 9. Create a GrandParent class with property_land. Ensure the GrandChild can "inherit" that property
class GrandParent:
    def __init__(self, property_land):
        self.property_land = property_land

class Parent(GrandParent):
    def __init__(self, property_land, property_house):
        super().__init__(property_land)
        self.property_house = property_house

class Child(Parent):
    def __init__(self, property_land, property_house, property_car):
        super().__init__(property_land, property_house)
        self.property_car = property_car

    def show(self):
        print(f"GrandParent Class : {self.property_land}")
        print(f"Parent Class : {self.property_house}")
        print(f"Child Class : {self.property_car}")

hierarchy = Child("50 acres of land", "Luxury Villa", "Range Rover")
hierarchy.show()
```

```python
# 10. The super() Keyword: Research the super() function. Use it in a child class to call a method from the parent class.
'''The super() function is a built-in Python tool used to call methods from a parent (or ancestor) class. It is most commonly used 
inside __init__ constructors to ensure that parent classes are properly initialized.'''

class Parent:
    def __init__(self):
        print("Parent initialized")

class Child(Parent):
    def __init__(self):
        super().__init__() 
        print("Child initialized")

obj = Child()
```

### Part C: Multiple Inheritance (Level 3)

```python
# 11. Skill Merger: Create a class Swimmer and a class Runner. Create a class Triathlete that inherits from both.
class Swimmer:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am a Swimmer")
        
class Runner:
    def __init__(self, **kwargs):
        # Forward call to the ultimate base class (object)
        super().__init__(**kwargs)
        print("I am a Runner")

# **kwargs - passes the remaining arguments to the next class

class Triathlete(Swimmer, Runner):
    pass

sports = Triathlete()
```

```python
# 12. The Diamond Problem: Create classes A, B, C, and D where B and C inherit from A, and D inherits from both B and C. (This is a classic "Top 1%" interview question).
class A:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class A.")

class B(A):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class B. I am inherited from Class A.")

class C(A):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class C. I am inherited from Class A.")

class D(B, C):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class D. I am inherited from Class B and C.")

inheritance = D()
```

```python
# 13. MRO (Method Resolution Order): Use the .mro() method on a class with multiple inheritance to see the order in which Python looks for methods.
class A:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class A.")

class B(A):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class B. I am inherited from Class A.")

class C(A):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class C. I am inherited from Class A.")

class D(B, C):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Class D. I am inherited from Class B and C.")

inheritance = D()
print(D.mro())
```

```python
# 14. Create a Camera class and a Phone class. Create a Smartphone class that inherits from both and has a method take_selfie().
class Camera:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am Camera Class")

class Phone:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am Phone Class")

class SmartPhone(Camera, Phone):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am SmartPhone Class")

    def take_selfie(self):
        print("I am taking a selfie.")

phone = SmartPhone()
```

```python
# 15. Conflict Handling: If Father and Mother classes both have a method eyes_color(), which one will the Child inherit first?
class Father:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Father Class.")

    def eyes_color(self):
        print("Inherited: Brown eyes from Father.")

class Mother:
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Mother Class.")

    def eyes_color(self):
        print("Inherited: Black eyes from Mother.")

class Child(Father, Mother):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        print("I am from Child Class")

child = Child()
child.eyes_color()
```

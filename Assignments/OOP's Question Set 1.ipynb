### Part B: Instance Methods & Logic (Level 2)

```python
# Bank Account: Create a class Account with balance. Add a method deposit(amount) that adds to the balance.
class Account():
    def __init__(self, initial_balance = 0.0):
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            print("Your amount is deposited successfully.")
            print("Your current balance is Rs.", self.balance + amount)
        else:
            print("Your amount cannot be deposited.. It must be greater than 0")

my_account = Account()
my_account.deposit(0)
```

```python
my_account.deposit(10000)
```

```python
# Area Calculator: Create a class Circle that takes radius. Add a method calculate_area() (use 3.14 * r * r).
class Circle():
    def __init__(self, r): 
        self.r = r
    def calculate_area(self):
        area = 3.14 * self.r * self.r
        print(f"The area of circle with radius {self.r} is {area}")

circle = Circle(5)
circle.calculate_area()
```

```python
# Status Checker: In a SmartDevice class, add a variable is_on = False. Create a method toggle() that switches the state.
class SmartDevice():
    def toggle(self, is_on = False):
        if is_on == False:
            self.is_on = True
            print(self.is_on)
        else:
            self.is_on = False
            print(self.is_on)

device = SmartDevice()
device.toggle()
```

```python
device.toggle(True)
```

```python
# String Representation: Research the __str__ method and implement it in your Laptop class so printing the object looks clean.
'''
The __str__ method is a built-in Python magic method that defines how an object should be represented as a readable string. When you use print(object) 
or str(object), Python automatically triggers this method under the hood.Without it, printing an object outputs a messy memory address like 
<__main__.Laptop object at 0x7f83b8>.
'''

class Laptop():
    def __init__(self, brand, ram, processor, price):
        self.brand = brand
        self.ram = ram
        self.processor = processor
        self.price = price

    def __str__(self):
        return f"Brand : {self.brand}\n Ram : {self.ram}\n Processor : {self.processor}\n Price : {self.price}"

mylaptop = Laptop("HP", "8GB", "intel i core", 45000)
print(mylaptop.__str__())
```

### Part C: Privacy & Encapsulation (Level 3)

```python
# Create a class Employee where salary is private (__salary).
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

emp = Employee("Sakshi", 12000)
emp.name
```

```python
# Getter Method: Add a method get_salary() to the Employee class that allows you to see the private salary safely.
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    def get_salary(self):
        return self.__salary
        
emp = Employee("Sakshi", 12000)
print(emp.name)
print(emp.get_salary())
```

```python
# Setter Method: Add a method set_salary(new_salary) that only updates the salary if the new_salary is greater than 0.
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    def set_salary(self, new_salary):
        if new_salary > 0:
            print(f"Initial salary : {self.__salary}")
            self.__salary = new_salary
            print("New salary is ", new_salary)
        
emp = Employee("Sakshi", 12000)
print(emp.name)
print(emp.set_salary(20000))
```

```python
# Create a protected variable _project_name in a class and access it from an object.
class Employee():
    def __init__(self, name, project_name):
        self.name = name
        self._project_name = project_name

new_project = Employee('Sakshi', 'QuizForge')
print(new_project.name)
print(new_project._project_name)
```

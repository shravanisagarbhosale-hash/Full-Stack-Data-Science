### Part B: Logic for NLP/Data Science (Level 2)

```python
# 6. Time Logger: Import the datetime module and create a function that returns the current time formatted for a log file.
from datetime import datetime
def return_current_time():
    now = datetime.now()
    return now.strftime("%Y-%m-%d %H:%M:%S")

print(f"[{return_current_time()}] INFO: Application started successfully.")
```

```python
# 7. Data Simulation: Use the random module to pick a random "sentiment" (Positive, Negative, Neutral) from a list.
import random
sentiment = ["Positive", "Negative", "Neutral"]
print(random.choice(sentiment))
```

```python
# 8. System Paths: Use os.path to check if a specific data file exists before trying to open it.
import os
filename = "data.txt"
if os.path.isfile(filename):
    print(f"{filename} found. Opening the file....")
    f = open(filename, 'r')
    data = f.read()
    print(data)
else:
    print(f"{filename} not found. Enter another file name ")
```

```python
# 9. String Constants: Import the string module and print all available punctuation marks (essential for cleaning NLP text).
import string
punctuation_marks = string.punctuation
print(punctuation_marks)
```

```python
# 10. JSON Handling: Use the json module to convert a Python dictionary into a JSON string.
import json
dictionary = {
    "Name" : "Shravani",
    "Education" : "Bsc CS",
    "Graduation Year" : 2026,
    "Current Course" : "Full Stack Data Science"
}

json_string = json.dumps(dictionary, indent=4)
print(json_string)
```

## 2. Local and Global Variables

```python
'''  Q2: Write a program where:
o You define a global variable counter with an initial value of 0.
o Create a function increment() that increments counter by 1 each time it is called and prints its value.
o Call the function three times and observe the output. '''

counter = 0

def increment():
    global counter
    counter += 1
    print(f"Counter value: {counter}")

increment()
increment()
increment()
```

```python
# Q3: Write a function demo_local() that declares a local variable message = "Hello, Local" and prints it. Outside the function, try to print 
#the same message variable and note the error.
def demo_local():
    message = "Hello, Local"
    print(f"Inside function: {message}")

demo_local()
print(f"Outside function: {message}")
```

```python
# Q4: Modify demo_local() from the previous question to access a global variable
# message = "Hello, Global" from within the function.
message = "Hello, Local"

def demo_local():
    global message
    print(f"Inside function: {message}")

demo_local()
print(f"Outside function: {message}")
```

## 3. Writing Docstrings

```python
'''
 Q5: Write a function greet(name) that takes a name as input and returns a greeting message like "Hello, John!".
o Add a proper docstring describing the purpose of the function, its input parameter, and its return value.
o Use help() to display the docstring. '''

def greet(name):
    """
    Generates a personalized greeting message.
    
    Parameters:
    name (str): The name of the person to greet.

    Returns:
    str: A greeting string in the format "Hello, <name>!".
    """
    return f"Hello, {name}!"

message = greet("John")
print(message)

print("\n--- Displaying Function Documentation ---")
help(greet)
```

## 4. Functions with Return

```python
#  Q7: Write a function is_even(n) that returns True if a number is even, otherwise returns False. Test the function with multiple inputs.
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

is_even(5)
```
